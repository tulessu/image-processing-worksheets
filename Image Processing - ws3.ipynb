```python
#first question - 8bit img to 32bit img
image = cv2.imread('/Users/tules/Downloads/eif8bit.jfif') #open the 8bit img
img = np.array(image, dtype=np.uint16) #change the type of the img
img = img * 256 #values in 16bit 
cv2.imwrite('16bit.png', img) #write the 16bit form to a file

img2 = np.float32(img) #convert to 32bit
img2 = img2 * 65536 #values in 32bit 
cv2.imwrite('32bit.png', img2) #write it to file

image2 = Image.open("32bit.png") 
#image2.show() #display the img

plt.imshow(image2)
plt.show()
```

```python
#second question - 32bit img to 64bit
img3 = np.float64(img) #convert to 64bit
img3 = img3 * 4294967296 #values in 64 bit 
cv2.imwrite('64bit.png', img3) #write

image3 = Image.open("64bit.png") 
#image3.show() 

plt.imshow(image3)
plt.show()
```
